```python
def starsLinearRegression(starsFeatures,starsTarget):
    '''Performs linear regression on the stars dataset and outputs data about results '''
    from sklearn.metrics import ConfusionMatrixDisplay, precision_score, recall_score, accuracy_score, explained_variance_score
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.model_selection import KFold 
    #standardize the features
    mms = MinMaxScaler(feature_range=(-1, 1))
    starsFeatures = mms.fit_transform(starsFeatures)
    
    #prepare k fold
    kf = KFold(n_splits=10)
    scores = np.zeros(10)
    variance = np.zeros(10)
    i = 0
    for train_idx, test_idx in kf.split(starsFeatures):
        features_train_kfold, features_test_kfold = starsFeatures[train_idx], starsFeatures[test_idx]
        labels_train_kfold, labels_test_kfold = starsTarget[train_idx], starsTarget[test_idx]
        #fit the model and generate predictions
        model = LinearRegression().fit(features_train_kfold, labels_train_kfold)
        model_pred = np.around(model.predict(features_test_kfold))
        scores[i] = accuracy_score(labels_test_kfold, model_pred)
        variance[i] = explained_variance_score(labels_test_kfold, model_pred)
        print(variance[i])
        i = i + 1

    print("average accuracy",scores.mean(), "standard deviation:",scores.std())
    print("variance:",variance.mean())
    
    #create example graph
    features_train, features_test, labels_train, labels_test = train_test_split(starsFeatures, starsTarget, test_size=0.25)
    
    model = LinearRegression(positive=False).fit(features_train, labels_train)
    
    # compute predictions for samples in the testing dataset, rounded to nearest integer
    model_pred = np.around(model.predict(features_test))
    ConfusionMatrixDisplay.from_predictions(labels_test, model_pred)
    plt.show()
    
```

```python
starsLinearRegression(starsFeatures,starsTarget)
```

```python
def garmentLinearRegression(garmentDataset, garmentTarget):
    '''Performs linear regression on the garment dataset and outputs information about results '''
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import MinMaxScaler
    from sklearn.metrics import mean_squared_error, explained_variance_score
    from sklearn.model_selection import KFold 
    #standardize the features
    mms = MinMaxScaler(feature_range=(-1, 1))
    garmentDataset = mms.fit_transform(garmentDataset)
    
    
    #prepare k fold
    kf = KFold(n_splits=10)
    scores = np.zeros(10)
    variance = np.zeros(10)
    i = 0
    for train_idx, test_idx in kf.split(garmentDataset):
        features_train_kfold, features_test_kfold = garmentDataset[train_idx], garmentDataset[test_idx]
        labels_train_kfold, labels_test_kfold = garmentTarget[train_idx], garmentTarget[test_idx]
        #fit the model and generate predictions
        model = LinearRegression().fit(features_train_kfold, labels_train_kfold)
        model_pred = model.predict(features_test_kfold)
        scores[i] = mean_squared_error(labels_test_kfold, model_pred)
        variance[i] = explained_variance_score(labels_test_kfold, model_pred)
        print("variance",variance[i])
        i = i + 1
    
    print("average error",scores.mean(), "standard deviation:",scores.std())
    print("variance:",variance.mean())
    
    #create example graph
    features_train, features_test, labels_train, labels_test = train_test_split(garmentDataset, garmentTarget, test_size=0.25)
    
    model = LinearRegression().fit(features_train, labels_train)

    
    # compute predictions for samples in the testing dataset
    model_pred = model.predict(features_test)


    # scatter plot of actual and predicted values for the target variable for first 10 samples
    plt.plot(labels_test[:10], 'o', label='Actual')
    plt.plot(model_pred[:10], 'o', label='Prediction')

    # add the legend to reflect which point denote which quantity
    plt.legend()
    plt.xlabel('Index of sample') # label of x-axis
    plt.ylabel('Actual/Predicted value of the target') # label of y-axis
```

```python
garmentLinearRegression(garmentDataset, garmentTarget)
```

Markdown question for 3_2:
   
I my belief, linear regression is not suitable for the stars dataset.

A significant reason for this is the fact that it is a classification problem with 3 possible labels. With 2 possible labels, it would be possible to use linear regression to find out which is more likely, however due to the fact that there are 3, this becomes unsuitable. The reason for this is because the labels have no continuous fashion inherent to them, instead being discrete and independent to one and other. However linear regression fails to see this, mistakingly seeing the labels as a continuous value. As a result, the integer value a label is assigned to randomly can dramatically effect its chance of being predicted.

We can see this clearly in the confusion graph made for the prediction of linear regression on the stars dataset. The label with integer value 1 is predicted by far the most, despite infact being the least common in the dataset. Meanwhile the label related to the value 2 is rarely ever predicted. The reason for this is that 0 and 2 are seen as the extreme values for prediction, while 1 is seen as the average between them.
